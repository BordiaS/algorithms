```python
## Combinations  https://leetcode.com/problems/combinations/

def find_combinations(n, k):
    
    result =[]
    N=[n]
    def helper(i,k, slate):
        if len(slate) ==k:
            result.append(slate.copy())
            return 
        if i<=N[0]:
            helper(i+1,k,slate)
            slate.append(i)
            helper(i+1,k,slate)
            slate.pop()
            
    helper(1,k,[])
    
    return result

find_combinations(5, 2)
```

## Permute


```python
class Solution:
    def helper(self,nums, i, res, partial):
        if i==len(nums):
            res.append(partial.copy())
            return 

        for j in range(i,len(nums)):
            #print(i, j)
            partial[i], partial[j]=  partial[j], partial[i]
            self.helper(nums, i+1, res, partial)
            partial[i], partial[j]=  partial[j], partial[i]

    def permute( self,nums):
        res = []
        self.helper(nums,0, res, nums)
        return res
```

```python
#Permute 2
import collections

def helper(nums,i, res, count, slate):
    if i == len(nums):
        res.append(slate.copy())
        return 
    
    for n in count:
        if count[n]>0:
            slate.append(n)
            count[n]-=1
            helper(nums,i+1, res, count, slate)
            count[n]+=1
            slate.pop()
            

def permute( nums):
    res = []
    count = collections.Counter(nums)
    helper(nums,0, res,count, [])
    return res

permute([2,1,1])
```

```python
def generate_all_combinations(arr, target):
    import collections
    
    res = []
    count = collections.Counter(arr)
    
    def helper(arr, target, count, slate, res):
        if sum(slate)==target:
            res.append(slate.copy())
            return 
        
        for n in count:
            if count[n] >0:
                
                slate.append(n)
                count[n]-=1
                helper(arr, target, count, slate, res)
                count[n]+=1
                slate.pop()
    
    helper(arr, target, count, [], res)
    return res

#     {
# "arr": [1, 1, 1, 1],
# "target": 2
# out : [[1,1]
# }


#     {
# "arr": [1, 2, 3 ],
# "target": 3
# out : [[1,2],[2,1],[3]]
# }

```
